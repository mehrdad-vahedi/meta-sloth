```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 1000
```

```python
def rand_ploter(x, y,_title,_color='blue'):
    plt.scatter(x, y, alpha=0.7,linewidths=0.1, color=_color)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(_title)
    plt.grid(True)
    # plt.axis('equal')
    plt.axis('scaled')
    plt.xlim(min(x), max(x))
    plt.ylim(min(y), max(y))
    plt.show()
```

تولید نمودار با اعداد رندم `numpy` برای نمونه

```python
arr = np.random.rand(N * 2)
arr = arr.reshape((N,2))

x = arr[:,0]
y = arr[:,1]

rand_ploter(x, y,"Uniform Distribution NumPy", "red")
```

```python
arr = np.random.randn(N * 2)
arr = arr.reshape((N,2))

x = arr[:,0]
y = arr[:,1]

rand_ploter(x, y,_title="Uniform Distribution NumPy",_color="green")
```
